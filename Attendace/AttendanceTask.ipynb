```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("20-03-2020.tsv",sep="\t")
```

```python
df1
```

```python
df1['Join & Leave Times'] = df1['Join & Leave Times'].str.split('-')
```

```python
df1
```

```python
df1["start"] = df1["Join & Leave Times"].apply(lambda x: x[0].replace('AM',''))
```

```python
df1["end"] = df1["Join & Leave Times"].apply(lambda x: x[1].replace('AM',''))
```

```python
df1.drop(["Join & Leave Times","Email Address"],axis=1,inplace=True)
```

```python
df1
```

```python
def hms_to_s(s):
    t = 0
    for i in s.split(':'):
        t = 60 * t + int(i)
    return t
#data = "10:14"
#hms_to_s(data)
#print(int(data)/60)
```

```python
df1["start"] = df1["start"].apply(hms_to_s)
```

```python
df1["end"] = df1["end"].apply(hms_to_s)
```

```python
df1["total-mins"] = (df1["end"]-df1["start"])
```

```python
df1['Name'] = df1.Name.str.upper().str.title()
```

```python
df1 = df1.groupby(['Name']).agg({'total-mins': 'sum'})
```

```python
df1
```
