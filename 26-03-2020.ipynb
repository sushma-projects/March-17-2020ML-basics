how
- inner - get common keys in both dataframes(intersection)
- outer - get keys atleast present in one dataframe(union)
- left - keys in the left dataframe
- right - keys in the right dataframe

```python
#storing the data in a csv file
total = pd.merge(data2,data1,left_on='Roll Number',right_on='No',how='outer')
total.to_csv("students.csv")
```
