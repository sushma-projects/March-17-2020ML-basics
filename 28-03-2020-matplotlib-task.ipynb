Draw a line using given axis values with suitable label in the x axis , y axis and a title.
<pre>
# x axis values
x = [1,2,3]
# y axis values
y = [2,4,1]
</pre>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# x axis values
x = [1,2,3]
# y axis values
y = [2,4,1]
plt.plot(x,y)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("Simple Graph of x and y values")
```

Plot two or more lines with  different widths and colors

```python
import numpy as np
x1 = np.random.randint(10,100,10)
y1 = np.random.randint(10,100,10)
x2 = np.random.randint(20,200,10)
y2 = np.random.randint(20,200,10)
plt.plot(x1,y1,color='red',linewidth='3')
plt.plot(x2,y2,color='green',linewidth='3')
plt.title("Plotting two or more lines with different widths and colors")
plt.show()
```

Plot two or more lines with different linestyles.

```python
help(plt.plot)
```

```python
x1 = np.random.randint(10,100,10)
y1 = np.random.randint(10,100,10)
x2 = np.random.randint(20,200,10)
y2 = np.random.randint(20,200,10)
plt.plot(x1,y1,color='red',linestyle='dashed')
plt.plot(x2,y2,color='green',linestyle='dotted')
plt.plot(x1*2,y1*2,color='yellow',linestyle='dashdot')
plt.title("Plotting two or more lines with different linestyles")
plt.show()
```

Plot two or more lines and set the line markers of your choice

```python
x1 = np.random.randint(10,100,10)
y1 = np.random.randint(10,100,10)
x2 = np.random.randint(20,200,10)
y2 = np.random.randint(20,200,10)
plt.plot(x1,y1,marker='^',color='red')
plt.plot(x2,y2,marker='*',color='green')
plt.title("Plotting two or more lines and set the line markers")
plt.show()
```

- Plot the data with given x and y values.
- Display only a part of it using xlim and ylim

```python
x = [1,2,3,4,5]
y = [2,4,9,16,25]
plt.plot(x,y,marker='*',color="red")
plt.xlim(0,6)
plt.ylim(0,30)
plt.xlabel("Plotting the x and y values using xlimit and ylimit")
plt.show()
```

Plot several lines with different format styles in one command using arrays.Use the plot function only once

```python
x = np.random.randint(1,10,10)
y = np.random.randint(2,20,10)
plt.plot(x,y,'r*',2*x,y,'g--',x*3,3*y,'y^',x*4,y*2,'b+')
```

Display a bar chart of the popularity of programming Languages. <br>
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
import pandas as pd
data = pd.DataFrame({'Programming languages': ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'], \
                                             'Popularity': [22.2, 17.6, 8.8, 8, 7.7, 6.7]})
data
```

```python
data['Programming languages'].value_counts()
```

```python
plt.bar(data['Programming languages'],data['Popularity'])
```

Display a bar chart of the popularity of programming Languages. Use different color for each bar

```python
plt.bar(data['Programming languages'],data['Popularity'],color='rgbkymc')
```

For the above data create a pie chart

```python
plt.pie(data['Popularity'],labels=data['Programming languages'],explode=(0, 0, 0, 0, 0,0.15))
```

For the above data create a pie chart mention your choice of colors with hexadecimal values

```python
plt.pie(data['Popularity'],labels=data['Programming languages'],\
        colors=["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076"])
```

Draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other<br>
use random.randn function

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x,y,facecolors='none', edgecolors='r')
```

draw a scatter plot using random distributions to generate circles of different sizes.

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = (np.random.sample(100) * 15) ** 2
plt.scatter(x,y, edgecolors='r',s=sizes,c=colors)
```
