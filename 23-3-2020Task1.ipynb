```python
import math
#If it is numerical column
print("datatype",data.user_id.dtypes)
#caluculate min,max,mean,mode,sum,std,median
print("min value",data.user_id.min())
print("max value",data.user_id.max())
print("mean value",data.user_id.mean())
print("sum",data.user_id.sum())
print("median",data.user_id.median())
mean = data.user_id.mean()
l = len(data.user_id)
sd = 0
for i in data.user_id:
    sd += (float(i)-mean)**2
sd = math.sqrt(sd/float(l-1))
print("std",sd)
#calculate 25%,50%,75% quartile values uisng np.quantile
np.quantile(data.user_id, [0.25,0.50,0.75], axis=0)
#calculate the range
print(data.min(),data.max())
```
